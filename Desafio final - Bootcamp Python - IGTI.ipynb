```
from sklearn.preprocessing import MinMaxScaler
normaliza = MinMaxScaler() # objeto para a normalização
```

```
entradas_normalizadas = normaliza.fit_transform(entradas)
```

```
saida = dataset[8]
```

```
from sklearn.model_selection import train_test_split
entradas_treino, entradas_teste, saida_treino, saida_teste = train_test_split(entradas_normalizadas,
                                                                              saida, test_size=0.30,
                                                                              random_state=42)
```

```
entradas_treino.shape, entradas_normalizadas.shape, entradas.shape
```

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
entradas_treino, entradas_teste, saida_treino, saida_teste = train_test_split(entradas_normalizadas,
                                                                              saida, test_size=0.30,
                                                                              random_state=42)

from sklearn.neighbors import KNeighborsClassifier

clf_KNN = KNeighborsClassifier(n_neighbors=5)
clf_KNN.fit(entradas_treino, saida_treino)

#realiza a previsão com os dados
saida_prevista = clf_KNN.predict(entradas_teste)

print()
acuracia = accuracy_score(saida_teste, saida_prevista)
print('Acurácia KNN: ', acuracia)
```

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
entradas_treino, entradas_teste, saida_treino, saida_teste = train_test_split(entradas_normalizadas,
                                                                              saida, test_size=0.30,
                                                                              random_state=42)

from sklearn.tree import DecisionTreeClassifier
clf_arvore = DecisionTreeClassifier(random_state=1)
clf_arvore.fit(entradas_treino, saida_treino)

#realiza a previsão com os dados
saida_prevista = clf_arvore.predict(entradas_teste)

print()
acuracia = accuracy_score(saida_teste, saida_prevista)
print('Acurácia arvore: ', acuracia)
```

```
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
entradas_treino, entradas_teste, saida_treino, saida_teste = train_test_split(entradas_normalizadas,
                                                                              saida, test_size=0.30,
                                                                              random_state=42)
from sklearn.neural_network import MLPClassifier
clf_mlp = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 10), random_state=1)
clf_mlp.fit(entradas_treino, saida_treino)

#realiza a previsão com os dados
saida_prevista = clf_mlp.predict(entradas_teste)

print()
acuracia = accuracy_score(saida_teste, saida_prevista)
print('Acurácia Redes Neurais MLP: ', acuracia)
```

```
import joblib

# salvando o modelo no disco
nome_do_arquivo = 'melhor_modelo.sav' #
melhor_modelo = clf_arvore
joblib.dump(melhor_modelo, nome_do_arquivo) 
```
